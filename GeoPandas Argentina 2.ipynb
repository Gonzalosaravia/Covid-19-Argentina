```python
def getPointCoords(row, geom, coord_type):
    """Calculates coordinates ('x' or 'y') of a Point geometry"""
    if coord_type == 'x':
        return row[geom].x
    elif coord_type == 'y':
        return row[geom].y
```

```python
gdf_provincias['x'] = gdf_provincias.apply(getPointCoords, geom='geometry', coord_type='x', axis=1)
gdf_provincias['y'] = gdf_provincias.apply(getPointCoords, geom='geometry', coord_type='y', axis=1)
gdf_provincias.head()
```

### Distancia entre dos lugares



```python
from geopy.distance import distance
import geopy
```

```python
Buenos_Aires = geo.geocode("Buenos Aires, Argentina")
Santa_Fe = geo.geocode("Santa Fe, Argentina")
```

```python
coord_BsAs = (Buenos_Aires.latitude, Buenos_Aires.longitude)
coord_SF = (Santa_Fe.latitude, Santa_Fe.longitude)

distancia = geopy.distance.distance(coord_BsAs, coord_SF )
print("La distancia entre Buenos Aires y Santa fe es de " + str(distancia))
```

### Crear función para poder calcular la distancia entre dos lugares cuales quiera.


```python
def distancia (origen, destino):
    lugar_origen = geo.geocode(origen)
    lugar_destino = geo.geocode(destino)
    
    coord_origen = (lugar_origen.latitude, lugar_origen.longitude)
    coord_destino = (lugar_destino.latitude, lugar_destino.longitude)
    
    dist = geopy.distance.distance(coord_origen, coord_destino)
    print("La distancia entre " + origen + " y " + destino + " es: " + str(dist))
    
    return
```
